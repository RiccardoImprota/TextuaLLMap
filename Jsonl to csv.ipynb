```python
root_dir = 'localdb'
output_dir = 'csv_formatted'

def process_jsonl_file(file_path):
    data_list = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            if line.strip():  # Skip empty lines
                json_obj = json.loads(line)
                data_list.append(json_obj)
    return data_list

def flatten_data(data_list):
    # Use pandas.json_normalize to flatten nested JSON structures
    df = pd.json_normalize(data_list, max_level=2)
    return df

for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith('.jsonl'):
            file_path = os.path.join(root, file)
            print(f'Processing file: {file_path}')

            # Compute the relative path from root_dir to the file
            rel_path = os.path.relpath(file_path, root_dir)

            # Change the extension from .jsonl to .csv
            rel_path = os.path.splitext(rel_path)[0] + '.csv'

            # Construct the output file path
            output_file = os.path.join(output_dir, rel_path)

            # Ensure the output directory exists
            os.makedirs(os.path.dirname(output_file), exist_ok=True)

            data_list = process_jsonl_file(file_path)
            df = flatten_data(data_list)
            df.to_csv(output_file, index=False)
            print(f'Data successfully written to {output_file}')
```
